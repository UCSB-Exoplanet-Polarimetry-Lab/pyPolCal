```python
from astropy.io import fits
import matplotlib.pyplot as plt
```

```python
from astropy.visualization import simple_norm
fn = '/home/thomasmc/charis_mm_testing/HD35532/cubes/data/raw/CRSA00155555_cube.fits'
hdul = fits.open(fn)
imagedata = hdul[1].data[10]
mask = imagedata >40000
snorm = simple_norm(imagedata,'log',)
plt.imshow(imagedata,origin='lower',norm=snorm, cmap='inferno')
plt.colorbar()
plt.plot(mask, origin='lower', cmap='gray', alpha=0.5)
```
