```python
import numpy as np
import matplotlib.pyplot as plt

# Define S1(theta1, phi) using the expanded formula
def S1(theta1_deg, phi_deg):
    theta1 = np.deg2rad(theta1_deg)
    phi = np.deg2rad(phi_deg)
    
    c1 = np.cos(2 * theta1)
    s1 = np.sin(2 * theta1)
    c2 = np.cos(2 * (theta1 + phi))
    s2 = np.sin(2 * (theta1 + phi))
    
    return -c2**2 * c1**2 - c2 * s2 * c1 * s1 + s2 * s1

# Compute I_out = 0.5 * (1 + S1)
theta1_vals = np.linspace(0, 180, 1000)
phi_fixed = 1
I_out_vals = 0.5 * (1 + S1(theta1_vals, phi_fixed))

# Plot
plt.figure()
plt.plot(theta1_vals, I_out_vals)
plt.xlabel(r'$\theta_1$ (degrees)')
plt.ylabel(r'$I_{\mathrm{out}}$')
plt.title(r'Output Intensity vs $\theta_1$ for $\phi = ' + str(phi_fixed) + r'^\circ$')
plt.grid(True)
plt.show()
```
