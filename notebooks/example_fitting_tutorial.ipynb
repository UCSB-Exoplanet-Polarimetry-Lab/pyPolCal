# Pseudo-Code Explanation of Fitting Routine

1. .csv file is read in and unnormalized double differences and sums are calculated
    - .csv files have single differences and sums (for each )
    - For multi-band imaging, the data for each wavelength is saved as a different numpy array
2. Configuration dictionaries are generated for each instrument state
    - This includes HWP angle and IMR angle
    - The default instrument configuration is saved with Wollaston 'o' beam and FLC state
