# Pseudo-Code Explanation of Fitting Routine

1. .csv file is read in: (pandas dataframe)
2. Configuration dictionaries are generated for each instrument state (list of dictionaries)
    - This includes HWP angle and IMR angle
    - The default instrument configuration is saved with Wollaston 'o' beam and FLC state A
3. From the saved dictionaries, normalized double differences and sums are calculated and combined into one array (numpy array)
    - .csv files have single differences and sums saved
    - For multi-band imaging, the data for each wavelength is saved as a different numpy array
4. scipy.minimize called with a likelihood function that takes in list of configurations and measured double differences and sums (?)
    - Double differences and sums from the model are calculated from the list of configurations
    - Float returned for the residuals between the model and the data
